```python
def plot_target(df,column,labels,title):
    df[column].value_counts(normalize=True).plot.bar()
    plt.xlabel(labels[0],fontdict = {'fontsize': 10, 'color': 'blue'})
    plt.ylabel(labels[1],fontdict = {'fontsize': 10, 'color': 'blue'})
    plt.title(title,fontdict = {'fontsize': 20, 'color': 'blue'})
    plt.show()

plot_target(clean_data,'TARGET',['Payment Difficulty','%'],'Clients with payment difficulties')
plot_target(Previous,'NAME_CONTRACT_STATUS',['Status','%'],'Previous Application Status')
```

```python

fulldf = pd.merge(left=clean_data,right=Previous,how='inner',on='SK_ID_CURR')
```

```python
fulldf.head()
```

```python
# Loan application status and Checking contract percentage
```

```python
fulldf['NAME_CONTRACT_STATUS'].value_counts(normalize=True)*100
```

```python

approved_df = fulldf[fulldf['NAME_CONTRACT_STATUS']=='Approved']
refused_df = fulldf[fulldf['NAME_CONTRACT_STATUS']=='Refused']
canceled_df=fulldf[fulldf['NAME_CONTRACT_STATUS']=='Canceled']
unused_df=fulldf[fulldf['NAME_CONTRACT_STATUS']=='Unused offer']
```

```python
# Analysis on basis 0f gender  'CODE_GENDER'

fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(20,5),sharey=True)
ax1=sns.countplot(ax=ax1,data=approved_df,x='CODE_GENDER')
ax1.set_title('Approved')
ax1.set_xlabel('CODE_GENDER')
ax1.set_ylabel('Number of Loans')
ax2=sns.countplot(ax=ax2,data=refused_df,x='CODE_GENDER')
ax2.set_title('refused')
ax2.set_xlabel('CODE_GENDER')
ax2.set_ylabel('Number of Loans')
ax3=sns.countplot(ax=ax3,data=canceled_df,x='CODE_GENDER')
ax3.set_title('canceled')
ax3.set_xlabel('CODE_GENDER')
ax3.set_ylabel('Number of Loans')
ax4=sns.countplot(ax=ax4,data=unused_df,x='CODE_GENDER')
ax4.set_title('unused_')
ax4.set_xlabel('CODE_GENDER')
ax4.set_ylabel('Number of Loans')
plt.tight_layout()
plt.show()
```

```python
# Maximum females applied for the loan and approved
# Females refuse more as compared to males
```
